```python
import matplotlib.pyplot as plt
import matplotlib.colors as cm
# from matplotlib.transforms import Bbox
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib widget
# Parameters for the torus
R = 2  # Major radius
r = 0.8  # Minor radius

# Create a meshgrid for u (angle around the main circle) and v (angle around the tube)
u1 = np.linspace(0, 2 * np.pi, 100)
v1 = np.linspace(0, 2 * np.pi, 50)
u1, v1 = np.meshgrid(u1, v1)

# Parametric equations for the torus
x1 = (R + r * np.cos(v1)) * np.cos(u1)
y1 = (R + r * np.cos(v1)) * np.sin(u1)
z1 = r * np.sin(v1)



u2 = np.linspace(-0.5*np.pi, 0.5 * np.pi, 100)
v2 = np.linspace(0, 2 * np.pi, 50)
u2, v2 = np.meshgrid(u2, v2)

# Parametric equations for the torus
x2 = (R + r * np.cos(v2)) * np.cos(u2)
y2 = (R + r * np.cos(v2)) * np.sin(u2)
z2 = r * np.sin(v2)






# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Torus surface
logocolor1='#40c4ff'
logocolor2='#007fc3'

shade=True
alpha=1
# ax.plot_surface(x1, y1, z1,color=logocolor1,edgecolor=None,shade=shade,lightsource=ls,alpha=alpha)
# ax.plot_surface(x2, y2, z2,color=logocolor1,shade=shade,lightsource=ls,alpha=alpha)

ls = cm.LightSource(10, 90)

ls1 = cm.LightSource(10, -90)

ls2 = cm.LightSource(10, -90)
shift=R/2

ax.plot_surface(-x2, y2-shift, z2, rstride=1, cstride=1, color=logocolor1, edgecolor=None, alpha=alpha,shade=shade,lightsource=ls1)


ax.plot_surface(z1, y1+shift, x1, rstride=1, cstride=1, color=logocolor1, edgecolor=None, alpha=alpha,shade=shade,lightsource=ls1)

ax.plot_surface(x2, y2-shift, z2, rstride=1, cstride=1, color=logocolor1, edgecolor=None, alpha=alpha,shade=shade,lightsource=ls)


# ax.plot_surface(z1, -y1+shift, x1, rstride=1, cstride=1, color='green', edgecolor=None, alpha=alpha,shade=shade,lightsource=ls,zsort='max')

# ax.plot_surface(x1, -y1-shift, z1, rstride=1, cstride=1, color='red', edgecolor=None, alpha=alpha,shade=shade,lightsource=ls,zsort='max')

# ax.plot_surface(x, z, y, rstride=1, cstride=1, color=logocolor2, edgecolor=None, alpha=alpha,shade=shade,lightsource=ls)

# # Add lines of latitude (constant v, varying u)
# for j in np.linspace(0, 2*np.pi, 12):  # meridians
#     x_line = (R + r * np.cos(j)) * np.cos(u[0])
#     y_line = (R + r * np.cos(j)) * np.sin(u[0])
#     z_line = r * np.sin(j) * np.ones_like(u[0])
#     ax.plot(x_line, y_line, z_line, color='darkblue', linewidth=1)

# # Add lines of longitude (constant u, varying v)
# for i in np.linspace(0, 2*np.pi, 12):  # parallels
#     x_line = (R + r * np.cos(v[:, 0])) * np.cos(i)
#     y_line = (R + r * np.cos(v[:, 0])) * np.sin(i)
#     z_line = r * np.sin(v[:, 0])
#     ax.plot(x_line, y_line, z_line, color='darkred', linewidth=1)

# Set aspect and remove axes
# ax.set_box_aspect([1+R/r, 1+R/r, 1])
ax.view_init(elev=45, azim=-20, roll=-10)
ax.set_aspect('equal')
ax.axis('off')
plt.show()

plt.savefig("logo_base.png",transparent=True,dpi=300)
```
